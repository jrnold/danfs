```python
import sqlite3
```

```python
con  = sqlite3.connect("dfas.sqlite3")
```

```python
c = con.cursor()
```

```python
import lxml.html
from lxml.etree import ParserError

import re
PATTERN = re.compile(r"^\((.*)\)")

r = c.execute("SELECT * FROM danfs_ships")
for row in r:
    try:
        html = lxml.html.fromstring(row[4])
        try:
            p1 = html.xpath("//p[1]")[0].text_content()
        except IndexError:
            continue
        m = PATTERN.match(p1)
        if m is not None:
            info = m.group(1)
            info = re.split("[;:]", info)
            print([x.strip() for x in info])
    except ParserError:
        # these are empty
        pass
```
